## Question 1

What is the difference between a NoSQL and a SQL database?  Which is better?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________
SQL databases are relational while NoSQL databases are non-relational or distributed.






## Question 2

Describe a situation where a NoSQL database might be a better choice than a relational database, and explain your reasoning.

Write your answer below this line: 
_______________________________________________________________________________________________________________________________
When you have a very large set of data and you need multiple computers to maintain the data. Or, if you need ensure that your data isn't destroyed if one database is destroyed, it is better to use a system that allows multiple databases.





## Question 3

Instead of structuring data into columns and rows, it structures it in json-like documents.







## Question 4

What is fault tolerance, and what does it have to do with Resilient Distributed Datasets?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________
Fault tolerance allows a system to continue operating if one or more parts fail. Systems that distribute data between multiple datasets, like Resilient Distributed Datasets, have a higher fault tolerance than centralized SQL dataasets.







## Question 5

What is MapReduce? How is it related to Hadoop?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________
MapReduce is an algorithm that scales data processing into multiple computing nodes by breaking down datasets into smaller tuples. It is one of the core components of the Hadoop system.
